```python
#modul untuk membuat tabel
from tabulate import tabulate

class Transaction:
    def __init__(self):
        self.items = list()
        """
        merupakan class parent yang akan digunakan untuk menginisiasi objek transaksi 
        dengan atribut items dengan menggunakan list kosong yang akan menyimpan atribut item
        yang akan dibeli.
        """

    def add_item(self, item):
        self.items.append(item)
        print(f'{item} telah berhasil ditambahkan ke keranjang')
        self.show_items()

        """
        Fungsi untuk menambahkan item ke dalam list items.
        berupa nama, jumlah dan harga yang dipesan oleh pembeli.
        """


    def update_item_name(self, old_name, new_name):
        for item in self.items:
            if item[0] == old_name:
                item[0] = new_name

        print(f'Item {old_name} telah berhasil diubah menjadi {new_name}')
        self.show_items()
        """
        Fungsi untuk mengganti nama item pada list items.
        old_name (string): Nama item yang akan diganti.
        new_name (string): Nama baru untuk item yang akan diganti.
        """

    def update_item_qty(self, name, qty):
        for item in self.items:
            if item[0] == name:
                item[1] = qty

        print(f'Jumlah {name} telah diubah menjadi {qty}')
        self.show_items()
        """
        Fungsi untuk mengganti jumlah item pada list items.
        name (string): Nama item yang ingin diubah jumlahnya.
        new_qty (int): Jumlah baru untuk item yang ingin diubah.
        """

    def update_item_price(self, name, price):
        for item in self.items:
            if item[0] == name:
                item[2] = price

        print(f'Harga {name} telah diubah menjadi {price}')
        self.show_items()
        """
        Fungsi untuk mengganti harga per item pada list items.
        name (string): Nama item yang ingin diubah harganya.
        new_price (int): Harga per item yang baru untuk item yang ingin diubah.
        """

    def delete_item(self, name):
        for item in self.items:
            if item[0] == name:
                self.items.remove(item)

        print(f'Anda telah menghapus : {name}') 
        self.show_items()
        """
        Fungsi untuk menghapus item dari list items.
        name (string): Nama item yang ingin dihapus dari list items.
        """

    def reset_transaction(self):
        self.items = []

        print('Anda telah mereset transaksi ini. Semua item telah dihapus.')
        """
        Fungsi untuk menghapus semua item dari list items.
        """

    def check_order(self):
        error = False
        for item in self.items:
            if (None, "", 0) in item or not all(item):
                error = True

        if error:
            print('Terdapat kesalahan input data')
        else:
            print('Pesanan Anda Sudah Benar')
            self.show_items()
        """
        Fungsi untuk mengecek apakah input data sudah benar atau tidak.
        """

    def show_items(self):
        show_items = []
        header = ['No.', 'Nama Item', 'Jumlah Item', 'Harga Satuan Item', 'Total Harga']
        show_items.append(header)

        n = 0
        for item in self.items:
          n += 1
          table_no = n
          name = item[0]
          qty = item[1]
          price = item[2]
          total = item[1] * item[2]
          item_data = [table_no, name, qty , price, total]
          show_items.append(item_data)


          print(tabulate(show_items, tablefmt = 'fancy_grid'))

        """
        Fungsi untuk menampilkan tabel nama item, jumlah item, harga satuan dan total harga
        """

    def total_price(self):
        total = 0
        for item in self.items:
            total += item[1] * item[2]

        print(f'Total harga belanjaan Anda Rp.{total}')
        if total > 500000:
            discount = total * 0.1
            print(f'Total diskon yang didapatkan adalah Rp.{discount}')
        elif total > 300000:
            discount = total * 0.08
            print(f'Total diskon yang didapatkan adalah Rp.{discount}')
        elif total > 200000:
            discount = total * 0.05
            print(f'Total diskon yang didapatkan adalah Rp.{discount}')
        else:
            discount = 0
            print('Anda tidak mendapat diskon')

        total_akhir = total - discount
        print(f'Total yang harus dibayar: {total_akhir}')
        """
        Fungsi untuk menghitung total harga dari seluruh item pada list items.
        Menghitung diskon dengan ketentuan yang telah ditetapkan sebelumnya.
        """
```
